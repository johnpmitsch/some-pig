```python
class Player:
    """
    Responsibilities:
    - Checks roll
    - Can roll again or hold
    - Base player rolls once per turn
    - Returns the score per turn
    
    Collaborators:
    - Can roll dice
    """
    def __init__(self, name="Player"):
        self.name = name
#         self.player_score = 0
    
    def roll(self):
        return random.randint(1,6)
    
    def hold(self):
        return 7
    
#     def check_roll(self, dice_value):
#         score = 0
#         if dice_value == 1:
#             return score
#         else:
#             score += dice_value
#         return score
    
    def turn(self, roll_x_times=1):
        turn_total = 0
        for _ in range(roll_x_times):
            score = self.roll()
            if score == 1:
                turn_total = 0
                return turn_total
            else:
                turn_total += score
                score = 0
        return turn_total
```

```python
class Game:
    """
    Responsibilities:
    - Keeps track of the turn total
    - Keeps track of the player score
    - Checks the outcome of the dice and gives a result
    - Keeps track of turns taken with 7 being the limit
    Collaborators:
    - A player plays the simulation
    """
    def __init__(self):
        self.turn_total = 0
        self.player_score = 0
        self.turns_remaining = 7
    
#     def update_turn_score(self, score):
#         return self.turn_total += score
        
#     def update_player_score(self, turn_score):
#         return self.player_score += turn_score

    def set_up_game(self):
        self.turn_total = 0
        self.player_score = 0
        self.turns_remaining = 7
    
    def play_game(self, player):
        self.set_up_game()
        while self.turns_remaining > 0:
            self.player_score += player.turn()
            self.turns_remaining -= 1
        return self.player_score
        
```

```python
class Three_roll_player(Player):
    """ This player rolls 3 times a turn and then holds"""
    def turn(self, roll_x_times=3):
        turn_total = 0
        for _ in range(roll_x_times):
            score = self.roll()
            if score == 1:
                turn_total = 0
                return turn_total
            else:
                turn_total += score
                score = 0
        return turn_total
```

```python
class Alice_player(Player):
    
    def turn(self):
        try:
            alice_num = random.randint(1, get_character_from_text(lower_case_alice)/4.5)
        except ValueError:
            alice_num = 5
            
        turn_total = 0
        for _ in range(alice_num):
                score = self.roll()
                if score == 1:
                    turn_total = 0
                    return turn_total
                else:
                    turn_total += score
                    score = 0
        return turn_total
```
